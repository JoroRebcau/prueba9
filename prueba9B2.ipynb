```python
#Se crea la clase elevator que tendrá por atributos las caracteristicas del problema
class elevator():
    stepC        : list
    initialStepC : int
    inputStepsC  : dict
    directionC   : str
    #Se crea la función del constructor
    def __init__(self, stepI, initialStepI, inputStepsI, directionI):
        self.stepC        = stepI
        self.initialStepC = initialStepI
        self.inputStepsC  = inputStepsI
        self.directionC  = directionI

    def getStepC(self):
        return self.stepC
    
    def getInitialStepC(self):
        return self.initialStepC

    def getInputStepsC(self):
        return self.inputStepsC

    def getDirectionC(self):
        return self.directionC

class building():
    elevator1    : elevator
    elevator2    : elevator

    def __init__(self, elevator1I, elevator2I):
        self.elevator1 = elevator1I
        self.elevator2 = elevator2I

    '''
    La siguiente función permite crear una lista con los valores posibles que seguirían en el proceso ya sea subiendo o bajando. 
    En caso de encontrar un valor que es menor que la posición actual (en subida) se borra y solo se tienen encuenta 
    los que son mayor. En caso de bajada lo contrario. Adicionalmente, verifica si ya no hay más opciones para seguir subiendo o bajando y 
    cambia la dirección
    '''
    def itIsMoreValues(self, direction, step, actualStep):
        stepsInclude = step.copy()
        for i in range(len(step)):
            if direction == "subiendo" and actualStep > step[i]:
                stepsInclude.remove(step[i])
            elif direction == "bajando" and actualStep < step[i]:
                stepsInclude.remove(step[i])
                    
        if direction == "subiendo" and stepsInclude == []:
            stepsInclude = step
            direction = "bajando"

        elif direction == "bajando" and stepsInclude == []:
            stepsInclude = step
            direction == "subiendo"
        return direction, stepsInclude
    '''
    El siguiente bloque encuentra el valor mayor o menor de entre todos los que se encuentran en la lista 
    debido a que gracias a la función anterior, no existiran en la lista ejemplo valores mayores que el actual si estamos en bajada
    o valores menores que el actual si estamos en subida, de modo que encontrando el mayor o menor sabremos cual es la siguiente parada.
    '''
    def moreClosedOne(self, stepsInclude):
        nextValue = stepsInclude[0]
        nextValue2 = 0
        if direction == "subiendo":
            for i in range(1,len(stepsInclude)):
                if nextValue > stepsInclude[i]: 
                        nextValue2 = nextValue
                        nextValue = stepsInclude[i] 
    
        elif direction == "bajando":
            for i in range(1,len(stepsInclude)):
                if nextValue < stepsInclude[i]:
                        nextValue2 = nextValue
                        nextValue = stepsInclude[i] 
        return nextValue, nextValue2

    '''
    La siguiente función tiene el código principal del algoritmo del ascensor
    '''
    def getRoute(self):
        #Se usan los atributos de la clase como variables locales del método
        step1 = self.elevator1.getStepC()
        actualStep1 = self.elevator1.getInitialStepC()
        inputSteps1 = self.elevator1.getInputStepsC()
        direction1 = self.elevator1.getDirectionC()

        step2 = self.elevator2.getStepC()
        actualStep2 = self.elevator2.getInitialStepC()
        inputSteps2 = self.elevator2.getInputStepsC()
        direction2 = self.elevator2.getDirectionC()

        step = list(set(step1 + step2))

        #Se informa del piso y la dirección actual
        print("Elevador 1 en piso: " + str(actualStep1))
        print("Elevador 1 " + direction1)

        print("Elevador 2 en piso: " + str(actualStep2))
        print("Elevador 2 " + direction2)

        #En el siguiente ciclo se va a implementar toda la interacción de los pisos ingresados, los ya grabados y el ascensor
        while True:
            #Se utiliza la función anterior para tener la dirección correcta y el vector de las opciones correcto
            direction, stepsInclude1 = self.itIsMoreValues(direction1, step, actualStep1)
            direction2, stepsInclude2 = self.itIsMoreValues(direction2, step, actualStep2)
            #Se utiliza la función anterior para tener los valores de los dos pisos mas cercanos apra ambos ascensores
            nextStep2,nexStep2_2 = self.moreClosedOne(stepsInclude1)
            nextStep1, nexStep1_2 = self.moreClosedOne(stepsInclude2)
            #Se calculan las distancias de cada ascensor a su piso mas cercano y al piso mas cercano del otro 
            firstDistance = actualStep1 - nextStep1
            secondDistance = actualStep2 - nextStep2
            thirdDistance = actualStep1 - nextStep2
            fourthDistance = actualStep2 - nextStep1
            #Se valora si la distancia del otro ascensor es menor que la del ascensor actual, de ser así,
            #se escoge al ascensor mas cercano a la distancia mas corta y el otro ascensor toma su segunda opción de más cercano
            if firstDistance > secondDistance and thirdDistance < secondDistance:
                nextStep1 = nextStep2
                nextStep2 = nexStep2_2
            elif firstDistance < secondDistance and fourthDistance < firstDistance:
                nextStep2 = nextStep1
                nextStep1 = nexStep1_2

            #Se imprimen la subida o bajada de piso dependiendo del caso
            if direction1 == "subiendo":
                for i in range(actualStep1+1, nextStep1):
                    print("Elevador en piso: " + str(i))
            elif direction1 == "bajando":
                for i in range(actualStep1-1-nextStep1):
                    print("Elevador en piso: " + str(actualStep1 -1 - i))
            
            print("Se movió el asensor 1")

            if direction1 == "subiendo":
                for i in range(actualStep1+1, nextStep1):
                    print("Elevador en piso: " + str(i))
            elif direction1 == "bajando":
                for i in range(actualStep1-1-nextStep1):
                    print("Elevador en piso: " + str(actualStep1 -1 - i))
            
            print("Se movió el asensor 2")
            
            #Se actualiza la posición actual al haber llegado
            actualStep1 = nextStep1
            actualStep2 = nextStep2
            print("Elevador 1 en piso: " + str(actualStep1))
            print("Elevador 1 en piso: " + str(actualStep2))
            #Se quita de la lista la opción debido a que se alcanzo el objetivo
            if actualStep1 in step:
                step.remove(actualStep1)
            if actualStep2 in step:
                step.remove(actualStep2)

            #Verificamos nuevamente la dirección y el estado de las opciones a seguir en la lista
            direction1, stepsInclude1 = self.itIsMoreValues(direction1, step, actualStep1)
            direction2, stepsInclude2 = self.itIsMoreValues(direction2, step, actualStep2)

            print("Elevador se detiene")

            #Se agregan los valores nuevos correspondientes que se van agregando gracias al diccionario 1
            for key, value in inputSteps1.items():
                if key == actualStep1:
                    if value not in step:
                        #Se guardan y se informa de que fueron agregados
                        step.append(value)
                        print("Piso ingresado " + str(step[-1]))
            
            #Se agregan los valores nuevos correspondientes que se van agregando gracias al diccionario 2
            for key2, value2 in inputSteps2.items():
                if key2 == actualStep2:
                    if value2 not in step:
                        #Se guardan y se informa de que fueron agregados
                        step.append(value2)
                        print("Piso ingresado " + str(step[-1]))

            #Al no haber mas opciones disponibles se acaba el movimiento y se rompa el ciclo
            if len(step) == 0:
                break
            #Se informa el sentido actual
            print("Elevador 1 " + direction1)
            print("Elevador 2 " + direction2)
        return 

step =  [5, 29, 13, 10]
initialStep = 4
inputSteps =  {5:2, 29: 10, 13: 1, 10:1}
direction = "subiendo"

step2 =  [6, 30, 14, 11]
initialStep2 = 31
inputSteps2 =  {6:3, 30:11, 14:2, 11:2}
direction2 = "bajando"

elevator1 = elevator(step, initialStep, inputSteps, direction)
elevator2 = elevator(step2, initialStep2, inputSteps2, direction2)
build1 = building(elevator1, elevator2)
build1.getRoute()
```
