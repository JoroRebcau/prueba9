```python
step =  [5, 29, 13, 10]
initialStep = 4
inputSteps =  {5:2, 29: 10, 13: 1, 10:1}
direction = "Subiendo"

def ascensor(step, initialStep, inputSteps, direction):
    actualStep = initialStep
    print("Elevador en piso: " + str(actualStep))
    #print("Elevador " + direction)
    while True:

        lessValue = step[0]
        for i in range(1, len(step)):
                if lessValue > step[i]:
                    lessValue = step[i] 
        nextStep = lessValue
        if actualStep < nextStep:
            direction = "subiendo"
            print("Elevador " + direction)
        else: 
            direction = "bajando"
            print("Elevador " + direction)

        if (actualStep - nextStep) > 0:
            for i in range(actualStep+1, nextStep):
                print("Elevador en piso: " + str(i))
        else:
            for i in range(nextStep, actualStep+1):
                print("Elevador en piso: " + str(i))
        actualStep = nextStep
        print("Elevador en piso: " + str(actualStep))
        step.remove(actualStep)

        for key, value in inputSteps.items():
            if key == actualStep:
                print("Elevador se detiene")
                step.append(value)
                print("Piso ingresado " + str(step[-1]))

        if len(step) == 0:
            break
    return 

ascensor(step,initialStep, inputSteps, direction)
```
